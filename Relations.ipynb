```python
import pandas as pd
import os
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')

l = pd.read_csv('l_clust.csv', index_col=[0])
b = pd.read_csv('b_clust.csv', index_col=[0])
j = pd.read_csv('j_clust.csv', index_col=[0])
l_mode = pd.read_csv('l_clust_mod.csv', index_col=[0])
b_mode = pd.read_csv('b_clust_mod.csv', index_col=[0])
```

# Correlation heatmap

**Federal Law**

```python
import matplotlib.pyplot as plt
import seaborn as sns
liste = b.columns.tolist()
liste.pop(1)
liste.pop(-1)
df = b[liste]
plt.figure(figsize=(12,10))
cor = df.corr(method='pearson').round(2)
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.savefig('corr_federal.png', bbox_inches="tight")
plt.show()
```

**State Law**

```python
liste = l.columns.tolist()
liste.pop(1)
liste.pop(-1)
df = l[liste]
plt.figure(figsize=(12,10))
cor = df.corr(method='pearson').round(2)
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.savefig('corr_state.png', bbox_inches="tight")
plt.show()
```

**Judicature**

```python
liste = j.columns.tolist()
liste.pop(1)
liste.pop(-1)
df = j[liste]
plt.figure(figsize=(12,10))
cor = df.corr(method='pearson').round(2)
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.savefig('corr_judicature.png', bbox_inches="tight")
plt.show()
```

# Cluster Distribution

```python
import re
l['Cluster'] = l['Cluster'].astype(str).replace(re.escape('0'), 'Cluster_0')
l['Cluster'] = l['Cluster'].astype(str).replace(re.escape('1'), 'Cluster_1')
l['Cluster'] = l['Cluster'].astype(str).replace(re.escape('2'), 'Cluster_2')
l[:] = l[:].astype(str).replace(re.escape('1'), 'Yes')
l[:] = l[:].astype(str).replace(re.escape('0'), 'No')

b['Cluster'] = b['Cluster'].astype(str).replace(re.escape('0'), 'Cluster_0')
b['Cluster'] = b['Cluster'].astype(str).replace(re.escape('1'), 'Cluster_1')
b['Cluster'] = b['Cluster'].astype(str).replace(re.escape('2'), 'Cluster_2')
b[:] = b[:].astype(str).replace(re.escape('1'), 'Yes')
b[:] = b[:].astype(str).replace(re.escape('0'), 'No')

j['Cluster'] = j['Cluster'].astype(str).replace(re.escape('0'), 'Cluster_0')
j['Cluster'] = j['Cluster'].astype(str).replace(re.escape('1'), 'Cluster_1')
j['Cluster'] = j['Cluster'].astype(str).replace(re.escape('2'), 'Cluster_2')
j[:] = j[:].astype(str).replace(re.escape('1'), 'Yes')
j[:] = j[:].astype(str).replace(re.escape('0'), 'No')


l_mode['Cluster'] = l_mode['Cluster'].astype(str).replace(re.escape('0'), 
                                                          'Cluster_0')
l_mode['Cluster'] = l_mode['Cluster'].astype(str).replace(re.escape('1'), 
                                                          'Cluster_1')
l_mode['Cluster'] = l_mode['Cluster'].astype(str).replace(re.escape('2'), 
                                                          'Cluster_2')
l_mode[:] = l_mode[:].astype(str).replace(re.escape('1'), 'Yes')
l_mode[:] = l_mode[:].astype(str).replace(re.escape('0'), 'No')

b_mode['Cluster'] = b_mode['Cluster'].astype(str).replace(re.escape('0'), 
                                                          'Cluster_0')
b_mode['Cluster'] = b_mode['Cluster'].astype(str).replace(re.escape('1'), 
                                                          'Cluster_1')
b_mode['Cluster'] = b_mode['Cluster'].astype(str).replace(re.escape('2'), 
                                                          'Cluster_2')
b_mode[:] = b_mode[:].astype(str).replace(re.escape('1'), 'Yes')
b_mode[:] = b_mode[:].astype(str).replace(re.escape('0'), 'No')
```
